```python
# Importando bibliotecas
from scripts.leitura_dados import LeituraDados
from scripts.min_max_dados import MinimoMaximoDados
import numpy as np
import matplotlib.pyplot as plt
```

```python
arquivo = "..\\dataset\\Andando-PausasNãoMarcadas.csv"

# Leitura dos dados de acordo com a extensão do arquivo
df = LeituraDados.resultado(arquivo)
```

```python
# Cálculo da norma Euclidiana para transformar os valores de aceleração em um número representativo
df['MAGNITUDE_ACEL'] = np.sqrt(df['X']**2 + df['Y']**2 + df['Z']**2)
```

```python
# Obtendo valor mínimo e máximo da magnitude de aceleração
valor_minimo = MinimoMaximoDados.resultado(df['MAGNITUDE_ACEL'])[0]
valor_maximo = MinimoMaximoDados.resultado(df['MAGNITUDE_ACEL'])[1]
```

```python
import seaborn as sns
def plot(df,start,end,showx=False,showy=False,showz=False,showTotal=True):
    legend_lst=[]
    if(showx):
        sns.lineplot(y = 'X', x = df.index[start:end], data = df[start:end])
        legend_lst+='X'
    if(showy):
        sns.lineplot(y = 'Y', x = df.index[start:end], data = df[start:end])
        legend_lst+='Y'
    if(showz):
        sns.lineplot(y = 'Z', x = df.index[start:end], data = df[start:end])
        legend_lst+='Z'
    if(showTotal):
        sns.lineplot(y = 'MAGNITUDE_ACEL', x = df.index[start:end], data = df[start:end])
        legend_lst+='MAGNITUDE_ACEL'
    plt.legend(legend_lst)
    plt.ylabel("magnitude")
    plt.title("Accelerometer Plot from "+str(start)+" till "+str(end), fontsize = 15)
    plt.show()
    
plot(df,0,df.index[-1])
```
